```python


# ╔═╗╦═╗╦ ╦╔╦╗  ╔═╗╔═╗╔═╗╔╗╔╔╦╗╔═╗  ╔═╗╦═╗╔═╗╔╦╗╦╔╦╗╔═╗   ╔═╗╦  ╦╔═╗╔╗╔╔╦╗╔═╗
# ║  ╠╦╝║ ║ ║║  ╠═╣║ ╦║╣ ║║║ ║ ║╣   ║  ╠╦╝║╣  ║║║ ║ ║ ║───║  ║  ║║╣ ║║║ ║ ║╣ 
# ╚═╝╩╚═╚═╝═╩╝  ╩ ╩╚═╝╚═╝╝╚╝ ╩ ╚═╝  ╚═╝╩╚═╚═╝═╩╝╩ ╩ ╚═╝   ╚═╝╩═╝╩╚═╝╝╚╝ ╩ ╚═╝
# Font Name: Calvin S
# http://patorjk.com/software/taag

#DEFINIR CLASE CLIENTE
import os
class Cliente:
    __nombre=""
    __direccion=""
    __calificacion_crediticia=""
    __ingresos=""
    
    def __init__(self, nombre="", direccion="", calificacion_crediticia="", ingresos=0):
        self.__nombre= nombre
        self.__direccion = direccion
        self.__calificacion_crediticia = calificacion_crediticia
        self.__ingresos = ingresos
    def __str__ (self):    
        return f"CLIENTE: \n\t nombre: {self.__nombre} \n\t dirección: {self.__direccion} \n\t calificación: {self.__calificacion_crediticia} \n\t ingresos: {self.__ingresos}"

    def ingresar_datos_por_terminal(self):
        self.__nombre= input("Nombre del cliente:")
        self.__direccion = input("Dirección:")
        self.__calificacion_crediticia = input("calificación crediticia:")
        self.__ingresos = input("ingresos del cliente:")
        print(f"DATOS IGRESADOS DEL CLIENTE: \n\t nombre: {self.__nombre} \n\t dirección: {self.__direccion} \n\t calificación: {self.__calificacion_crediticia} \n\t ingresos: {self.__ingresos}")
        
    def Create_cliente(self):
        try:
            path="./Cliente.txt"
            cliente_abierto = open(path,"w+")
            cliente_abierto.write("\n"+self.__nombre + "|" + self.__direccion + "|"
                                  +self.__calificacion_crediticia + "|" + str(self.__ingresos))
            cliente_abierto.close()
            print(f"CLIENTE REGISTRADO EN ARCHIVO")
        except Exception as error:
            print(f"No se pudo crear el cliente: {error}")
            
    def Read_cliente(self):
        try:
            path="./Cliente.txt"
            cliente_abierto = open(path)
            lineas  = cliente_abierto.readlines()
            cliente_abierto.close()
            print("CLIENTE OBTENIDO DE ARCHIVO")
            print(lineas[len(lineas)-1])
        except Exception as error:
            print(f"No se pudo obtener el cliente el archivo: {error}")
            
    def Update_cliente(self):
        self.__nombre = input("Escriba nuevo nombre: ")
        self.__direccion = input("Escriba nueva dirección ")
        self.__calificacion_crediticia = input("Escriba nueva calificación crediticia")
        self.__ingresos = input("Escriba nuevos ingresos")
        
        try:
            path="./Cliente.txt"
            cliente_abierto = open(path,"w")
            cliente_abierto.write("\n"+self.__nombre + "|" + self.__direccion + "|"+self.__calificacion_crediticia + "|" + self.__ingresos)
            cliente_abierto.close()
            print(f"CLIENTE ACTUALIZADO EN ARCHIVO")
        except Exception as error:
            print(f"No se pudo actualizar el cliente: {error}")
            
    def Delete_cliente(self):
        try:
            path="./Cliente.txt"
            cliente_abierto = open(path, "r+")
#             if 
            os.remove(path)

            print(f"ARCHIVO DE CLIENTE ELIMINADO")
        except Exception as error:
            print(f"no se puede eliminar el archivo: {error}")
            
            

    def menu_cliente(self):
        
        opciones ={
            "1":self.Create_cliente,
            "2":self.Read_cliente,
            "3":self.Update_cliente,
            "4":self.Delete_cliente
        }
        
        seleccion=""
        while seleccion !="0":
            seleccion=input("Seleccione la opción: \n Nuevo:1 \n Buscar 2 \n Actualizar: 3 \n Eliminar 4:\n Terminar:0 ")
            print(f"opcion seleccionada {seleccion}")
            if(seleccion!="0"):
                opciones[seleccion]()
        
        
        
        
cliente1= Cliente("marco","yaruqui- 0e56", "bueno", 800.000)

cliente1.menu_cliente()


```
