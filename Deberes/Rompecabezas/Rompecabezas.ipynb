```python
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
from PIL import Image

# mapache = misc.face()

class Rpc():
 
    #fs, fi = np.vsplit(mapache,2)
    #fsc1, fsc2 = np.hsplit(fs,2)
    #fic1, fic2 = np.hsplit(fi,2)

    pfsc1 = None
    pfsc2 = None
    pfic1 = None
    pfic2 = None
    
    # Guardar lasposiciones
    def __init__(self, aimagen):
        fs, fi = np.vsplit(aimagen,2)
        fsc1, fsc2 = np.hsplit(fs,2)
        fic1, fic2 = np.hsplit(fi,2)
        self.pfsc1 = fic1
        self.pfsc2 = fsc1
        self.pfic1 = fsc2
        self.pfic2 = fic2
        self.desordenar();
        
    # un rompecabezas puede
    # imprimir piezas
    # mover piezas
    def imprimir(self): 
        fs=np.append(self.pfsc1,self.pfsc2,axis=1)
        fi= np.append(self.pfic1,self.pfic2, axis=1)
        image=np.append(fs,fi,axis=0)
        plt.imshow(image)
    
    def mover(self):
        pi= input("posicin inicial: ")
        pf= input("posicin final:")
        piezas={
            "1":self.pfsc1,
            "2":self.pfsc2,
            "3":self.pfic1,
            "4":self.pfic2,
        }

        aux = piezas[pf]
        piezas[pf]=piezas[pi]
        piezas[pi]= aux

        self.pfsc1=piezas["1"]
        self.pfsc2=piezas["2"]
        self.pfic1=piezas["3"]
        self.pfic2=piezas["4"]
        self.imprimir()
    def desordenar(self):
        pass

imagen_jpg = Image.open('dragfsand.jpg')
imagen_array = np.array(imagen_jpg)
np.shape(imagen_array)

# plt.imshow(imagen_array)
rpc = Rpc(imagen_array)
rpc.imprimir()
```
